### Ex1

- Write a function that adds sequence numbers to each item in a list
```
add_sequence(['a', 'b', 7]) ---> [(1, 'a'), (2, 'b'), (3, 7)]
```

````python
# Your code here
def add_sequence_v1(L):
    myiter=iter(range(1,len(L)+1))
    return list(map(lambda x:(next(myiter),x),L))

add_sequence_v1(['a', 'b', 7])
````

````python
def add_sequence_v2(L):
    result=[]
    for i in range(len(L)):
        result.append((i+1, L[i]))
    return result

add_sequence_v2(['a', 'b', 7])
````

````python
def add_sequence_v3(L):
    return [(i+1,L[i]) for i in range(len(L))]

add_sequence_v3(['a', 'b', 7])
````

````python
def add_sequence_v4(L):
    result=[]
    for i, x in enumerate(L):
        result.append((i+1,x))
    return result

add_sequence_v4(['a', 'b', 7])
````

````python
def add_sequence_v5(L):
    return [(i+1,x) for i,x in enumerate(L)]

add_sequence_v5(['a', 'b', 7])
````

### Ex4
Write function to add 'ing' at the end of a given string (length should be at least 3). 

If the given string already ends with 'ing' then add 'ly' instead. If the string length of the given string is less than 3, leave it unchanged.

"speak" → "speaking"

"willing" → "willingly"

````python
# Your code here
def add_ing(txt):
    if len(txt)<3:
        return txt
    elif txt[-3:]=='ing':
        return txt+'ly'
    else:
        return txt+'ing'

print(add_ing('speak'))
print(add_ing('willing'))
print(add_ing('do'))
````

### Ex5
Write a python program that prints all odd numbers between 1 and N

````python
# Your code here
def print_odds_v1(N):
    for i in range(1,N+1):
        if i%2!=0:
            print(i)

print_odds_v1(10)
````

````python
def print_odds_v2(N):
    for i in range(1,N+1,2):
        print(i)
        
print_odds_v2(10) 
````

````python
def print_odds_v3(N):
    print(list(range(1,N+1,2)))
        
print_odds_v3(10) 
````

```
Your company has a new Stocktaking system. In order to run the process accurately, two employees will do the same process separately. They use barcode scanner to scan all items in the inventory then all barcode numbers of all scanned items are saved in text file.
This process generates two data files (one file from each employee): invdata1.txt and invdata2.txt.

```
- Write code to find the number of items scanned more than one time by mistake (Duplicates) in each file.
- Write code that find barcode numbers found in file1 not file2 and vice versa.


````python
#### Your code here
f=open('invdata1.txt','r')
content=f.read()
f.close()

items=content.split(',')
print('The number of duplicates is: ',len(items)- len(set(items)))
````

````python
f=open('invdata2.txt','r')
content=f.read()
f.close()

items=content.split(',')
print('The number of duplicates is: ',len(items)- len(set(items)))
````

````python
def find_number_of_duplicates(filename):
    f=open(filename)
    content=f.read()
    f.close()
    items=content.split(',')
    print('The number of duplicates in',filename,'is: ',len(items)- len(set(items)))
````

````python
find_number_of_duplicates('invdata1.txt')
find_number_of_duplicates('invdata2.txt')
````
