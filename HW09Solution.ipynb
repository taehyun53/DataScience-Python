Write code to create a Pandas DataFrame, 'Employee', whose columns are 'Name', 'Service_Years' and 'Department'. The function should extract the data from the file (emp.txt).

```python
# Your code here
import pandas as pd
employee=pd.read_csv('emp.txt', sep=' ', header=None)
del employee[0]
del employee[2]
del employee[3]
del employee[4]
del employee[5]
del employee[7]
del employee[8]
del employee[10]
employee.columns=['Name','Department','Service_Years']
employee
```

```python
# Your code here
import pandas as pd
employee=pd.read_csv('emp.txt', sep=' ', header=None)
for col_name in [0,2,3,4,5,7,8,10]:
    del employee[col_name]
employee.columns=['Name','Department','Service_Years']

employee
```

```python
import pandas as pd
f=open('emp.txt')
lines=f.readlines()
f.close()
list_of_names=[l.split()[1] for l in lines]
list_of_departments=[l.split()[6] for l in lines]
list_of_years=[int(l.split()[-2]) for l in lines]

employee=pd.DataFrame({'Name':list_of_names, 'Service_Years':list_of_years, 'Department':list_of_departments})
employee
```
