### Ex1

- Write a function that adds sequence numbers to each item in a list
```
add_sequence(['a', 'b', 7]) ---> [(1, 'a'), (2, 'b'), (3, 7)]
```

````python
def add_sequence_v2(L):
    result = []
    for i in range(len(L)):
        result.append((i+1, L[i]))
    return result

add_sequence_v2(['a','b',7])
````

````python
def add_sequence_v4(L):
    result = []
    for i, x in enumerate(L):
        result.append((i+1, x))
    return result

add_sequence_v4(['a','b',7])
````

````python
i = 0
the_sum = 0

while True:
    if i % 7 == 0 and i < 100:
        the_sum += i
        i += 1
    if i > 100:
        break

print(True)

````
